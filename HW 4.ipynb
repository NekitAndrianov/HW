```python
df = pd.read_csv('flights.csv')
df.head()
```

```python
df.DEPARTURE_DELAY
```

```python

df_cringe = df[['ARRIVAL_DELAY',"AIRLINE", "ORIGIN_AIRPORT"]].groupby(by=["ORIGIN_AIRPORT", "AIRLINE"],as_index = False).mean()
df_cringe.loc[df_cringe["ARRIVAL_DELAY"] > 25].sort_values(by = "ARRIVAL_DELAY", ascending=False).head(50)
```

Видно, что у большей части связок авиакомпания - аэропорт среднее время задержки не превосходит получаса. Только у 33 среднее время превышает час. Большая часть этих аэропортов не имеет трехбуквенной классификации, что говорит, что это небольшие региональные аэропорты
