```python
def find_highest(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        a = find_highest(lst[1:])
        if a < lst[0]:
            return lst[0]
        else:
            return a
        
```

```python
Test.assert_equals(find_highest([8]), 8)
Test.assert_equals(find_highest([-1, 3, 5, 6, 99, 12, 2]), 99)
Test.assert_equals(find_highest([0, 12, 4, 87]), 87)
```

```python
def is_disarium(n):
    
    def length_calculator(num):
        length = 0
        while (num != 0):
            length += 1
            num = num // 10
        return length
    
    length = length_calculator(n)
    s = 0
    numberForUsage = n
    for i in range(length):
        digit = numberForUsage % 10
        s += digit**(length - i)
        numberForUsage = numberForUsage // 10
    if s == n:
        return True
    else:
        return False
    
```

```python
def num_grid(lst):
    i = 0
    field = []
    for i in range (len(lst)):
        field.append([])
        for j in range (len(lst[0])):
            field[i].append(0)
    for i in range (len(lst)):
        for j in range (len(lst[0])):
            if lst[i][j] == "#":
                if (i != 0 and j !=0):
                    field[i-1][j-1] +=1
                if i != 0 :
                    field[i-1][j] +=1
                if j !=0:
                    field[i][j-1] +=1
                if i != 0 and j != len(lst[0])-1:
                    field[i-1][j+1] +=1
                if j != len(lst[0])-1:
                    field[i][j+1] +=1
                if i != len(lst)-1 and j != len(lst[0])-1:
                    field[i+1][j+1] +=1
                if i != len(lst)-1 :
                    field[i+1][j] +=1
                if i != len(lst)-1 and j != 0 :
                    field[i+1][j-1] +=1
    for i in range (len(lst)):
        for j in range (len(lst[0])): 
            if lst[i][j] == "#":
                field[i][j] = "#"
            else:
                field[i][j] = str(field[i][j])
    return field
```

```python
def poker_hand_ranking(hand):
    suit = []
    rank = []
    for i in hand:
        if len(i) == 3:
            suit.append(i[2])
            rank.append(10)
        else:
            if i[0] == 'J':
                suit.append(i[1])
                rank.append(11)
            if i[0] == 'Q':
                suit.append(i[1])
                rank.append(12)
            if i[0] == 'K':
                suit.append(i[1])
                rank.append(13)
            if i[0] == 'A':
                suit.append(i[1])
                rank.append(14)
            if i[0] != 'J' and i[0] != 'Q' and i[0] != 'K' and i[0] != 'A':
                suit.append(i[1])
                rank.append(int(i[0]))
    if royal_flush_check (rank, suit):
        return "Royal Flush"
    if straight_flush_check (rank, suit):
        return "Straight Flush"
    if four_check (rank, suit):
        return "Four of a Kind"
    if full_house_check (rank, suit):
        return "Full House"
    if flush_check (rank, suit):
        return "Flush"
    if straight_check (rank, suit):
        return "Straight"
    if set_check (rank, suit):
        return "Three of a Kind"
    if two_pair_check (rank, suit):
        return "Two Pair"
    if pair_check (rank, suit):
        return "Pair"
    return "High Card"
            
        
    
def pair_check (rank, suit):
    if len(set(rank)) != 5:
        return True
    return False

def two_pair_check (rank, suit):
    if len(set(rank)) == 3:
        if rank.count(rank[0]) == 3 or rank.count(rank[1]) == 3 or rank.count(rank[2]) == 3:
            return False
        return True
    return False

def set_check (rank, suit):
    if len(set(rank)) == 3:
        if rank.count(rank[0]) == 3 or rank.count(rank[1]) == 3 or rank.count(rank[2]) == 3:
            if rank.count(rank[0]) != 2 and rank.count(rank[1]) != 2 and rank.count(rank[2]) != 2 and rank.count(rank[3]) != 2:
                return True
    return False

def straight_check (rank, suit):
    if len(set(rank)) == 5:
        rank.sort()
        for i in range (4):
            if rank[i+1] != rank[i]+1:
                return False
        return True
    else:
        return False
    
def flush_check (rank, suit):
    if len(set(suit)) == 1:
        return True
    return False

def full_house_check (rank, suit):
    if len(set(rank)) == 2:
        if rank.count(rank[0]) != 4 and rank.count(rank[1]) != 4:
            return True
    return False


def four_check (rank, suit):
    if rank.count(rank[0]) == 4 or rank.count(rank[1]) == 4:
        return True
    return False

def straight_flush_check (rank, suit):
    if len(set(suit)) == 1:
        if len(set(rank)) == 5:
            rank.sort()
            for i in range (4):
                if rank[i+1] != rank[i]+1:
                    return False
            return True
    return False

def royal_flush_check (rank, suit):
    rank.sort()
    if rank[4] != 14:
        return False
    else:
        if len(set(suit)) == 1:
            if len(set(rank)) == 5:
                rank.sort()
                for i in range (4):
                    if rank[i+1] != rank[i]+1:
                        return False
                return True
    return False
```

```python
Test.assert_equals(poker_hand_ranking(["10h", "Jh", "Qh", "Ah", "Kh"]), "Royal Flush")
Test.assert_equals(poker_hand_ranking(["3h", "5h", "Qs", "9h", "Ad"]), "High Card")
Test.assert_equals(poker_hand_ranking(["10s", "10c", "8d", "10d", "10h"]), "Four of a Kind")
Test.assert_equals(poker_hand_ranking(["4h", "9s", "2s", "2d", "Ad"]), "Pair")
Test.assert_equals(poker_hand_ranking(["10s", "9s", "8s", "6s", "7s"]), "Straight Flush")
Test.assert_equals(poker_hand_ranking(["10c", "9c", "9s", "10s", "9h"]), "Full House")
Test.assert_equals(poker_hand_ranking(["8h", "2h", "8s", "3s", "3c"]), "Two Pair")
Test.assert_equals(poker_hand_ranking(["Jh", "9h", "7h", "5h", "2h"]), "Flush")
Test.assert_equals(poker_hand_ranking(["Ac", "Qc", "As", "Ah", "2d"]), "Three of a Kind")
Test.assert_equals(poker_hand_ranking(["Ad", "Kd", "Qd", "Jd", "9d"]), "Flush")
Test.assert_equals(poker_hand_ranking(["10h", "Jh", "Qs", "Ks", "Ac"]), "Straight")
Test.assert_equals(poker_hand_ranking(["3h", "8h", "2s", "3s", "3d"]), "Three of a Kind")
Test.assert_equals(poker_hand_ranking(["4h", "Ac", "4s", "4d", "4c"]), "Four of a Kind")
Test.assert_equals(poker_hand_ranking(["3h", "8h", "2s", "3s", "2d"]), "Two Pair")
Test.assert_equals(poker_hand_ranking(["8h", "8s", "As", "Qh", "Kh"]), "Pair")
Test.assert_equals(poker_hand_ranking(["Js", "Qs", "10s", "Ks", "As"]), "Royal Flush")
Test.assert_equals(poker_hand_ranking(["Ah", "3s", "4d", "Js", "Qd"]), "High Card")
```
