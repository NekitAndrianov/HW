```python
fig, ax = plt.subplots(figsize=(15, 10))
width = 0.5 
HarePercent = Hares/(Hares+Lynxes)*100
LynxPercent = Lynxes/(Hares+Lynxes)*100
ax.bar(Years, HarePercent, width, label = 'Hares')
ax.bar(Years, LynxPercent, width, bottom = HarePercent,
       label='Lynxes')

ax.set_ylabel('%', fontsize=15)
ax.set_xlabel('Year', fontsize=15)
ax.set_title('Ratio of hares and lynxes')
ax.legend()
plt.xticks(np.arange(1900, 1921, step=1))

plt.show()
```
